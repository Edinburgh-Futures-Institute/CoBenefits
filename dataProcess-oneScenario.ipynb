#### Add LADs to the SEF table from the lookup LAD tables

```python
df_lad_NI = pd.read_csv("static/LAD/NI_DZ_LAD.csv")
df_lad_Eng = pd.read_csv("static/LAD/Eng_Wales_LSOA_LADs.csv")
df_lad_Scotland = pd.read_csv("static/LAD/Scotland_DZ_LA.csv", encoding='latin1')
```

```python
df_socio
```

```python
df_lad_Eng.head()
```

```python
df_lad_NI.head()
```

```python
df_lad_Scotland.head()
```

```python
lookup_EN = dict(zip(df_lad_Eng['LSOA11CD'], df_lad_Eng['LAD22CD']))
lookup_EN_2 = dict(zip(df_lad_Eng['LSOA21CD'], df_lad_Eng['LAD22CD']))

# lookup_NI = dict(zip(df_lad_NI['DZ2021_code'], df_lad_NI['DEA2014_code']))
lookup_NI = dict(zip(df_lad_NI['DZ2021_code'], df_lad_NI['LGD2014_code']))

lookup_SCO = dict(zip(df_lad_Scotland['DZ2011_Code'], df_lad_Scotland['LA_Code']))

for idx, row in df_socio[df_socio.LSOA_DZ_CD.str.startswith("E")].iterrows():
    lsoa = row['LSOA_DZ_CD']
    if lsoa in lookup_EN:
        df_socio.at[idx, 'LAD'] = lookup_EN[lsoa]
    elif lsoa in lookup_EN_2:
        df_socio.at[idx, 'LAD'] = lookup_EN_2[lsoa]
    # else:
    #     print(row)

for idx, row in df_socio[df_socio.LSOA_DZ_CD.str.startswith("W")].iterrows():
    lsoa = row['LSOA_DZ_CD']
    if lsoa in lookup_EN:
        df_socio.at[idx, 'LAD'] = lookup_EN[lsoa]
    elif lsoa in lookup_EN_2:
        df_socio.at[idx, 'LAD'] = lookup_EN_2[lsoa]
        
for idx, row in df_socio[df_socio.LSOA_DZ_CD.str.startswith("N")].iterrows():
    lsoa = row['LSOA_DZ_CD']
    if lsoa in lookup_NI:
        df_socio.at[idx, 'LAD'] = lookup_NI[lsoa]
        
for idx, row in df_socio[df_socio.LSOA_DZ_CD.str.startswith("S")].iterrows():
    lsoa = row['LSOA_DZ_CD']
    if lsoa in lookup_SCO:
        df_socio.at[idx, 'LAD'] = lookup_SCO[lsoa]
        
```
